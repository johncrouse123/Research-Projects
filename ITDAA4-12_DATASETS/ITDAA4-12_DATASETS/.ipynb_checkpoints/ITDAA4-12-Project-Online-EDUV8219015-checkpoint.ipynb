```python
# Question 1

# - Combine the data from all three years into a single dataset 
# by selecting only the relevant columns and adding an indicator column 
# to specify the year of each survey.

print("\n" + "-" * 60 + "\n")

# Ensure all columns are displayed
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)

# Connect to the database
conn = sqlite3.connect("github.db")
cur = conn.cursor()

# Define the columns needed
columns_needed = [
    "YearsCode", "MainBranch", "Country", "EdLevel", "LanguageHaveWorkedWith",
    "LanguageWantToWorkWith", "DatabaseHaveWorkedWith", "DatabaseWantToWorkWith",
    "Age"
]

# Load and clean each year's data
def load_data(table_name, year):
    df = pd.read_sql_query(f"SELECT {', '.join(columns_needed)} FROM {table_name}", conn)
    df["Year"] = year
    return df

# Load data from each year
data_2021 = load_data("data_2021", 2021)
data_2022 = load_data("data_2022", 2022)
data_2023 = load_data("data_2023", 2023)

# Combine all three datasets
combined_df = pd.concat([data_2021, data_2022, data_2023], ignore_index=True)

# Load and display the first 10 rows of the new table
print("\nPreview of records from combined_data:\n")
df_combined = pd.read_sql_query("SELECT * FROM combined_data LIMIT 10", conn)
print(df_combined)
print("\n" + "-" * 60 + "\n")
```

```python
# - Standardise the Country column, ensuring consistency in naming 
# for countries with variations

# Country standardisation mapping
country_mapping = {
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "Russian Federation": "Russia",
    "South Korea": "Korea, South",
    "Democratic People's Republic of Korea": "North Korea",
    "Côte d'Ivoire": "Ivory Coast",
    "Iran, Islamic Republic of...": "Iran",
    "Syrian Arab Republic": "Syria",
    "Venezuela, Bolivarian Republic of...": "Venezuela",
    "Palestine": "State of Palestine",
    "Eswatini": "Swaziland",  # Official name change from Swaziland to Eswatini
    "Bosnia and Herzegovina": "Bosnia & Herzegovina",
    "Hong Kong (S.A.R.)": "Hong Kong",
    "Republic of Korea": "South Korea",
    "Republic of Moldova": "Moldova",
    "The former Yugoslav Republic of Macedonia": "North Macedonia",
    "Democratic Republic of the Congo": "Congo, Democratic Republic of the",
    "Czech Republic": "Czechia",
    "Viet Nam": "Vietnam",
    "United Arab Emirates": "UAE",
    "Israel": "State of Israel",
    "Ivory Coast": "Côte d'Ivoire",
    "Brunei Darussalam": "Brunei",
    "Kyrgyzstan": "Kyrgyz Republic",
    "Tajikistan": "Tajikistan (Republic of)",
    "Turkmenistan": "Turkmenistan (Republic of)",
    "Macedonia": "North Macedonia",
    "Honduras": "Honduras (Rep. of)",
    "Slovak Republic": "Slovakia",
    "Sao Tome and Principe": "Sao Tome and Príncipe",
    "Timor-Leste": "East Timor",
    "Mongolia": "Mongolia (People's Republic of)",
    "Macau": "Macau (S.A.R.)",
    "Myanmar": "Burma",
    "Laos": "Lao People's Democratic Republic",
    "Eswatini": "Swaziland",  # Official name change from Swaziland to Eswatini
    "Vietnam": "Viet Nam",  # Possible case for consistency
    "Brunei Darussalam": "Brunei",
    "The Gambia": "Gambia",
    "Congo": "Congo, Republic of the"
}

# Standardise the 'Country' column using the mapping
combined_df['Country'] = combined_df['Country'].replace(country_mapping)

# Filter out None values from the 'Country' column and sort the country names
country_list = sorted(combined_df['Country'].dropna().unique())

print("List of unique countries in alphabetical order:\n")
for country in country_list:
    print(country)
print("\n" + "-" * 60 + "\n")

# Save the combined data back into the database as a new table
combined_df.to_sql("combined_data", conn, if_exists="replace", index=False)
```

```python
# - Handle missing data and subsetting the dataset 
# to include a manageable selection of countries 
# to facilitate clear visualizations and analysis. 
# Justify your choices for missing value treatment and country selection.

# Load the combined data
combined_df = pd.read_sql_query("SELECT * FROM combined_data", conn)

# Handle missing data
# Seperate numerical and categorical columns
numerical_cols = combined_df.select_dtypes(include = ['int64', 'float64']).columns
categorical_cols = combined_df.select_dtypes(include = ['object']).columns

# Fill numerical missing values with the median
for col in numerical_cols:
    median = combined_df[col].median()
    combined_df[col] = combined_df[col].fillna(median)

# Fill categorical missing values with the mode
for col in categorical_cols:
    mode = combined_df[col].mode()
    if not mode.empty:
        combined_df[col] = combined_df[col].fillna(mode[0])

# Subset top 10 countries
top_countries = combined_df['Country'].value_counts().nlargest(10).index
print("Top 10 countries by number of responses:\n")
print(combined_df['Country'].value_counts().nlargest(10).to_string())
print("\n" + "-" * 60 + "\n")

subset_df = combined_df[combined_df['Country'].isin(top_countries)]

# Justification for handling missing data:
# - Numerical columns were filled with the median 
# because the median is less affected by extreme values or outliers, 
# providing a more robust representation of central tendency.
# - Categorical columns were filled with the mode 
# to maintain the most common category, which ensures 
# that the overall categorical distribution is preserved.

# Justification for country selection:
# - The top 10 countries by response count were selected 
# to reduce complexity and allow for more focused, 
# clear visualisations and analyses. 
# These countries also provide enough data 
# for statistically meaningful comparisons.

# Overwrite the cleaned and subsetted data
subset_df.to_sql("combined_data", conn, if_exists = "replace", index = False)
```

```python
# - Categorize the values in the MainBranch column 
# into meaningful groups to provide a clear understanding of respondent roles.

# Define the mapping for categorising MainBranch values
mainbranch_mapping = {
    "I am a student who is learning to code": "Student",
    "I am a developer by profession": "Developer",
    "I used to be a developer by profession, but no longer am": "Former Developer",
    "I code primarily as a hobby": "Hobbyist",
    "I am not primarily a developer, but I write code sometimes as part of my work": "Occasional Developer",
    "I am not primarily a developer, but I write code sometimes as part of my work/studies": "Occasional Developer",
    "I am learning to code": "Learner"
}
    
# Apply the mapping to the MainBranch  column
combined_df['MainBranch'] = combined_df['MainBranch'].map(mainbranch_mapping)

# Update the MainBranch column based on the mapping
for original_value, new_value in mainbranch_mapping.items():
    cur.execute("""
                UPDATE combined_data
                SET MainBranch = ?
                WHERE MainBranch = ?
                """, (new_value, original_value))
    
# Commit the changes
conn.commit()

# Fetch and display all unique EdLevel values
cur.execute('SELECT DISTINCT MainBranch FROM combined_data')
unique = cur.fetchall()

print("MainBranch values after categorisation:\n")
for mainbranch in unique:
    print(mainbranch[0])
print("\n" + "-" * 60 + "\n")
```

```python
# - Reclassify the EdLevel column into broader educational categories 
# to create concise, interpretable groupings.

# Define the mapping for reclassifying the EdLevel values
edlevel_mapping = {
    "Bachelor’s degree (B.A., B.S., B.Eng., etc.)": "Undergraduate Degree",
    "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)": "Graduate Degree",
    "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)": "High School",
    "Some college/university study without earning a degree": "Some College",
    "Other doctoral degree (Ph.D., Ed.D., etc.)": "Doctoral Degree",
    "Something else": "Other",
    "Primary/elementary school": "Elementary School",
    "Associate degree (A.A., A.S., etc.)": "Associate Degree",
    "Professional degree (JD, MD, etc.)": "Professional Degree",
    "Professional degree (JD, MD, Ph.D, Ed.D, etc.)": "Professional Degree"
}

# Update the EdLevel column using the mapping
for old_value, new_value in edlevel_mapping.items():
    cur.execute('''
        UPDATE combined_data
        SET EdLevel = ?
        WHERE EdLevel = ?
    ''', (new_value, old_value))

# Commit the changes
conn.commit()

# Fetch and display all unique EdLevel values
cur.execute('SELECT DISTINCT EdLevel FROM combined_data')
unique = cur.fetchall()

print("EdLevel values after reclassification:\n")
for edlevel in unique:
    print(edlevel[0])
print("\n" + "-" * 60 + "\n")
```

```python
# 3. Relationship Analysis
#  3.1. How does YearsCode correlate with the use of the top 5 databases and programming languages 
#  currently in use?

# Convert YearsCode to numeric
def convert_years(x):
    if pd.isna(x):
        return np.nan
    elif isinstance(x, str):
        if 'less' in x.lower():
            return 0.5
        elif 'more' in x.lower():
            return 50
        else:
            try:
                return float(x)
            except:
                return np.nan
    return x

df = combined_df.copy()

df['YearsCode'] = df['YearsCode'].apply(convert_years)

# Top technologies
top_databases = ['PostgreSQL', 'MySQL', 'SQLite', 'MongoDB', 'Microsoft SQL Server']
top_languages = ['JavaScript', 'HTML/CSS', 'Python', 'SQL', 'TypeScript']

# Explode the multi-response columns
df_databases = df[['DatabaseHaveWorkedWith', 'YearsCode']].dropna()
df_databases['DatabaseHaveWorkedWith'] = df_databases['DatabaseHaveWorkedWith'].str.split(';')
df_databases = df_databases.explode('DatabaseHaveWorkedWith')

df_langs = df[['LanguageHaveWorkedWith', 'YearsCode']].dropna()
df_langs['LanguageHaveWorkedWith'] = df_langs['LanguageHaveWorkedWith'].str.split(';')
df_langs = df_langs.explode('LanguageHaveWorkedWith')

# Filter only the top 5
df_databases = df_databases[df_databases['DatabaseHaveWorkedWith'].isin(top_databases)]
df_langs = df_langs[df_langs['LanguageHaveWorkedWith'].isin(top_languages)]

# Plotting the boxplot for YearsCode by Database
plt.figure(figsize = (10, 6))
sns.boxplot(data = df_databases, x = 'DatabaseHaveWorkedWith', y = 'YearsCode')
plt.title('Years of Experience by Top 5 Databases Used')
plt.xlabel('Database')
plt.ylabel('Years of Coding Experience')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# Plotting the boxplot for YearsCode by Programming Languege
plt.figure(figsize = (10, 6))
sns.boxplot(data = df_langs, x = 'LanguageHaveWorkedWith', y = 'YearsCode')
plt.title('Years of Experience by Top 5 Programming Languages Used')
plt.xlabel('Programming Language')
plt.ylabel('Years of Coding Experience')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# Question 3

# 3.1. Preprocess your data as necessary and perform sentiment analysis on the content column 
# to categorize user reviews as positive, neutral, or negative. 
# Save the sentiments for each review in a new column called “Sentiment”.

# Load the dataset
df = pd.read_csv("chatgpt_reviews.csv")

# Function to clean and preprocess text
def preprocess_text(text):
    if pd.isnull(text):
        return ""
    return str(text).encode('ascii', 'ignore').decode('utf-8')  # ASCII-encode to remove emojis

# Overwrite the original 'content' column with cleaned values
df['content'] = df['content'].apply(preprocess_text)

# Function to classify sentiment based on polarity
def classify_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    if polarity > 0.1:
        return 'Positive'
    elif polarity < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

# Add sentiment column
df['Sentiment'] = df['content'].apply(classify_sentiment)

# View sample output
print(df[['content', 'Sentiment']].head())
print("\n" + "-" * 60 + "\n")
```

```python
# 3.2. Using visualizations, highlight features or issues that are frequently mentioned 
# in both positive and negative reviews. Additionally, identify which aspects users seem to love 
# about ChatGPT vs which features they dislike.

# Download stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Function to clean and tokenise text
def clean_text(text):
    text = str(text).lower()
    text = ''.join([char for char in text if char not in string.punctuation])
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]
    return " ".join(tokens)

# Clean reviews
df['content'] = df['content'].apply(clean_text)

# Seperate by sentiment
positive_reviews = df[df['Sentiment'] == 'Positive']['content']
negative_reviews = df[df['Sentiment'] == 'Negative']['content']

# CountVectoriser to extract top words
def get_top_n_words(corpus, n = None):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis = 0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)
    return words_freq[:n]

# Get top 35 words
top_pos_words = get_top_n_words(positive_reviews, 35)
top_neg_words = get_top_n_words(negative_reviews, 35)

# Convert to DataFrame for plotting
pos_df = pd.DataFrame(top_pos_words, columns = ['word', 'freq'])
neg_df = pd.DataFrame(top_neg_words, columns = ['word', 'freq'])

# Plot bar charts
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.barplot(x='freq', y='word', hue='word', data=pos_df, palette='Greens_r', legend=False)
plt.title('Top Words in Positive Reviews')

plt.subplot(1, 2, 2)
sns.barplot(x='freq', y='word', hue='word', data=neg_df, palette='Reds_r', legend=False)
plt.title('Top Words in Negative Reviews')

plt.tight_layout()
plt.show()
```

```python
# Set number of top words to display
top_n = 50

# Get top words for positive and negative reviews
top_positive_words = Counter(' '.join(positive_reviews).split()).most_common(top_n)
top_negative_words = Counter(' '.join(negative_reviews).split()).most_common(top_n)

# Top Words in Positive Reviews
print("\nTop Words in Positive Reviews:\n")
for word, freq in top_positive_words:
    print(f"{word:<15} : {freq}")
```

```python
# Top Words in Negative Reviews
print("\n\nTop Words in Negative Reviews:\n")
for word, freq in top_negative_words:
    print(f"{word:<15} : {freq}")
```

```python
# 3.3. Critically discuss how user satisfaction scores have changed over time. 
# Motivate your answer using a line plot.

# Make sure date is datetime type
df['date'] = pd.to_datetime(df['date'])

# Group by month and calculate average score
df_monthly = df.groupby(df['date'].dt.to_period('M')).agg({'score': 'mean'}).reset_index()
df_monthly['date'] = df_monthly['date'].dt.to_timestamp()

# Plotting
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_monthly['date'], df_monthly['score'], marker='o', linestyle='-', color='blue')
plt.title('Average User Satisfaction Score Over Time')
plt.xlabel('Date')
plt.ylabel('Average Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Round the score for cleaner output
df_monthly['score'] = df_monthly['score'].round(2)

# Rename columns for clarity
df_monthly.rename(columns = {'score': 'Average Score', 'date': 'Month'}, inplace = True)

# Print the results
print("Monthly Average User Satisfaction Scores:\n")
print(df_monthly.to_string(index = False))
```
